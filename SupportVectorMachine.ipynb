```python

import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
plt.style="seaborn"
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load DataFrames from CSV
X_train_pca_df = pd.read_csv("X_train_pca.csv")
X_val_pca_df = pd.read_csv("X_val.csv")
X_test_pca_df = pd.read_csv("X_test.csv")
Y_train = pd.read_csv("Y_train.csv")
Y_val = pd.read_csv("Y_val.csv")
Y_test = pd.read_csv("Y_test.csv")
```

```python
# for the optimisation function as explained in class;our gradient wrt w and b yields:
# gradient = 2 * self.lambda_param * self.weights - x_i * y[idx]

# using gradient descent 
# self.weights -= self.learning_rate * gradient
# self.bias -= self.learning_rate * y[idx]
```

```python


class SVM:
    def _init_(self, learning_rate=0.01, lambda_param=0.01, num_epochs=100):
        self.learning_rate = learning_rate
        self.lambda_param = lambda_param
        self.num_epochs = num_epochs
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # Initialize weights and bias
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Gradient Descent
        for epoch in range(self.num_epochs):
            for idx, x_i in enumerate(X):
                condition = y[idx] * (np.dot(x_i, self.weights) - self.bias) >= 1
                if condition:
                    self.weights -= self.learning_rate * (2 * self.lambda_param * self.weights)
                else:
                    gradient = 2 * self.lambda_param * self.weights - x_i * y[idx]
                    self.weights -= self.learning_rate * gradient
                    self.bias -= self.learning_rate * y[idx]
            print(f"Epoch {epoch + 1}/{self.num_epochs}: Completed")

    def predict(self, X):
        return np.sign(np.dot(X, self.weights) - self.bias)
```

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train_pca_df, Y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_val_pca_df)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(Y_val, y_pred))
```

```python
y_pred_test = clf.predict(X_test_pca_df)
```

```python
# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(Y_test, y_pred_test))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.metrics import f1_score

# Define a range of C values
C_values = [0.01, 0.1, 1, 10, 100]  # You can adjust this range as needed

# Initialize lists to store F1 scores
f1_scores = []

# Train the model with different values of C
for C_value in C_values:
    # Create a svm Classifier with specified C value
    clf = svm.SVC(kernel='linear', C=C_value)

    # Train the model using the training sets
    clf.fit(X_train_pca_df, Y_train)

    # Predict the response for the validation dataset
    y_pred = clf.predict(X_val_pca_df)

    # Calculate F1 score and append to list
    f1 = f1_score(Y_val, y_pred)
    f1_scores.append(f1)

# Plot F1 scores versus C values
plt.figure(figsize=(8, 6))
plt.plot(C_values, f1_scores, marker='o')
plt.xscale('log')
plt.xlabel('C')
plt.ylabel('F1 Score')
plt.title('F1 Score vs. C')
plt.grid(True)

# Save the figure
plt.savefig('f1_score_vs_C.png')
plt.show()
```
